```python
import os
gt_path = '/home/hls/workspace/creat_holiday_label/perfect_result.dat'
lab_path = '/home/hls/workspace/creat_holiday_label/label'
query_list = []
with open (gt_path,'r') as f:
    for line in f.readlines():
        query = line.split(' ')[0][:-4]
        query_list.append(query)
print(query_list)
print(len(query_list))
if not os.path.exists(lab_path):
    os.makedirs(lab_path)
for q in query_list:
    q_file = os.path.join(lab_path,q+'_query.txt')
    with open (q_file,'w') as f:
        f.write(q)
```
